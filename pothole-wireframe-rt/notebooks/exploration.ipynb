# Exploration Notebook

This notebook is intended for exploratory data analysis (EDA) of the dataset used for the real-time pothole detection system. The goal is to understand the characteristics of the data, visualize the distribution of potholes, and prepare the data for training the model.

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import os

# Set the path to the dataset
data_path = '../data/raw/'

# Function to load images from the dataset
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder, filename))
        if img is not None:
            images.append(img)
    return images

# Load images
images = load_images_from_folder(data_path)

# Display some sample images
plt.figure(figsize=(12, 8))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.axis('off')
plt.show()
```
